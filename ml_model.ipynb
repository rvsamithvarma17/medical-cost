```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
file_path = '/content/drive/MyDrive/Medical_Cost_Prediction.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
print("Dataset Info:")
print(data.info())
```

```python
data['BMI'] = data['BMI'].fillna(data['BMI'].mean())
data['Annual_Income'] = data['Annual_Income'].fillna(data['Annual_Income'].median())
data['Medical_Expenses'] = data['Medical_Expenses'].fillna(data['Medical_Expenses'].mean())
```

```python
print("Dataset Info:")
print(data.info())
```

```python
print("\nMissing Values After Handling:")
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['gender_encoded'] = label_encoder.fit_transform(data['Gender'])
print("\nDataset after Label Encoding:")
data
```

```python
print(data.columns)
```

```python
y = data['Medical_Expenses']
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data['scaled_Annual_income'] = scaler.fit_transform(data[['Annual_Income']])
data['scaled_Medical_Expenses'] = scaler.fit_transform(data[['Medical_Expenses']])
data
```

```python
x = data.drop(['Gender','Patient_ID','Medical_Expenses','Annual_Income','Medical_Expenses'],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=53)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
lr_results= pd.DataFrame(['Linear regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','Training MSE','Training R2','Testing MSE','Testing R2']
lr_results
```
